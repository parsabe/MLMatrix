# Pearson correlation coefficient

<p>
The Pearson correlation coefficient, often represented by the letter <strong>r</strong>, is a statistical measure that quantifies the strength and direction of the linear relationship between two variables. It ranges from <strong>-1</strong> to <strong>+1</strong>, with values closer to <strong>0</strong> indicating a weak relationship, values close to <strong>+1</strong> indicating a strong positive relationship, and values close to <strong>-1</strong> indicating a strong negative relationship.
</p>

<h2>Key Aspects</h2>
<ul>
<li>
  <h3>Linearity</h3>
  <p>
    Pearson's correlation measures the strength of a linear relationship between two variables. It is suitable for data where the relationship can be approximated by a straight line.
  </p>
</li>
<li>
  <h3>Strength of Relationship</h3>
  <p>
    The magnitude of the correlation coefficient indicates the strength of the relationship. Values closer to <strong>1</strong> or <strong>-1</strong> suggest a stronger relationship, while values closer to <strong>0</strong> suggest a weaker relationship.
  </p>
</li>
<li>
  <h3>Direction of Relationship</h3>
  <p>
    The sign of the correlation coefficient indicates the direction of the relationship. A positive value indicates that as one variable increases, the other tends to increase, while a negative value indicates that as one variable increases, the other tends to decrease.
  </p>
</li>
<li>
  <h3>Limitations</h3>
  <p>
    Pearson's correlation does <em>not</em> imply causation and cannot be used for non-linear relationships or when the data is not normally distributed. In those cases, other correlation methods, such as Spearman's rank correlation, may be more appropriate.
  </p>
</li>
</ul>

<h2>Formula</h2>
<p>
The formula for calculating the Pearson correlation coefficient is based on the covariance of the two variables divided by the product of their standard deviations:
</p>


$$
r \;=\;
\frac{\displaystyle \sum_{i}(x_{i} - \mu_{x})(y_{i} - \mu_{y})}
{\displaystyle \sum_{i}(x_{i} - \mu_{x})^{2} \;\times\; \sum_{i}(y_{i} - \mu_{y})^{2}}
$$


<h2>Interpretation</h2>
<p>
Correlation coefficients are often categorized as weak, moderate, or strong based on their value. For example:
</p>
<ul>
<li>0.7 to 1.0: Strong positive correlation</li>
<li>0.3 to 0.5: Weak to moderate positive correlation</li>
<li>-0.3 to -0.5: Weak to moderate negative correlation</li>
<li>-0.7 to -1.0: Strong negative correlation</li>
</ul>






<h1>Area Under the Curve (AUC)</h1>
<p>
In machine learning, the Area Under the Curve (AUC) is a performance metric used to evaluate binary classification models. It quantifies how well a model can distinguish between positive and negative classes across all possible classification thresholds. The AUC is calculated by measuring the area under the Receiver Operating Characteristic (ROC) curve.
</p>

<h2>Key Aspects of AUC</h2>
<ul>
<li>
<h3>Binary Classification</h3>
<p>
AUC is primarily used for evaluating binary classification tasks, where the goal is to predict one of two outcomes (e.g., yes/no, positive/negative).
</p>
</li>
<li>
<h3>ROC Curve</h3>
<p>
The ROC curve is a graphical plot that shows the trade-off between the True Positive Rate (TPR) and the False Positive Rate (FPR) at different classification thresholds.
</p>
</li>
<li>
<h3>AUC Value</h3>
<p>
AUC is a single number representing the area under the ROC curve. It ranges from <strong>0</strong> to <strong>1</strong>, with higher values indicating better model performance.
</p>
</li>
<li>
<h3>Perfect Model</h3>
<p>
A perfect model, which perfectly separates positive and negative classes, would have an AUC of <strong>1.0</strong>.
</p>
</li>
<li>
<h3>Random Guessing</h3>
<p>
A model that performs no better than random guessing would have an AUC of <strong>0.5</strong>.
</p>
</li>
<li>
<h3>Interpretation</h3>
<p>
A higher AUC indicates a better model's ability to distinguish between positive and negative classes, even when the classification threshold changes.
</p>
</li>
<li>
<h3>Threshold Independence</h3>
<p>
AUC provides a general measure of performance that is not tied to any specific classification threshold.
</p>
</li>
</ul>

<h2>Formula</h2>
<p>
The AUC is computed as the integral of the True Positive Rate over the False Positive Rate:
</p>
<p class="formula">

$$\mathrm{AUC} \;=\; \int_{0}^{1} \mathrm{TPR}(t)\,\mathrm{d}\mathrm{FPR}(t)$$

</p>

<h2>Summary</h2>
<p>
In essence, AUC provides a comprehensive measure of a model‚Äôs ability to distinguish between classes across all thresholds, making it a valuable tool for comparing and evaluating different binary classification models.
</p>




<h1>Receiver Operating Characteristic (ROC) Curve</h1>
<p>
A Receiver Operating Characteristic (ROC) curve is a graphical representation that helps evaluate the performance of a binary classification model. It plots the true positive rate (sensitivity) against the false positive rate (1‚Äìspecificity) at various threshold settings, showing how well a model distinguishes between positive and negative cases as the threshold changes.
</p>

<h2>Purpose</h2>
<p>
ROC curves are used to assess the accuracy and effectiveness of a diagnostic test or classification model. They help illustrate the trade-off between correctly identifying positive cases (high sensitivity) and incorrectly labeling negatives as positives (low false positive rate).
</p>

<h2>How It‚Äôs Created</h2>
<p>
The ROC curve is generated by varying the threshold for classifying an instance as positive. For each threshold, compute:
</p>
<p class="formula">

$$\text{TPR (Sensitivity)} = \frac{\text{TP}}{\text{TP} + \text{FN}}, \quad
\text{FPR} = \frac{\text{FP}}{\text{FP} + \text{TN}} = 1 - \text{Specificity}$$

</p>
<p>
Then plot TPR versus FPR for all thresholds from 0 to 1.
</p>

<h2>Interpreting the Curve</h2>
<ul>
<li>
    A curve closer to the top-left corner indicates better performance (high TPR with low FPR).
</li>
<li>
    A curve near the diagonal (45¬∞) line suggests the model has little discriminative ability, akin to random guessing.
</li>
</ul>

<h2>Area Under the Curve (AUC)</h2>
<p>
The AUC is the area under the ROC curve, summarizing performance in a single number. An AUC of 1.0 indicates a perfect model, while an AUC of 0.5 indicates no discriminative power.
</p>

<h2>Applications</h2>
<ul>
<li><strong>Medical Diagnosis:</strong> Evaluating the accuracy of diagnostic tests.</li>
<li><strong>Machine Learning:</strong> Comparing and validating classification models.</li>
<li><strong>Signal Detection:</strong> Balancing detection sensitivity against false alarms.</li>
</ul>




## CHI-Squared method

The Chi-Square Test is a statistical method used to check how independent a feature is from the target variable.

    üí° In short:
    ‚ÄúDoes this feature help predict the target?‚Äù

It‚Äôs commonly used for:

    Classification problems ‚úÖ

    Categorical target variables ‚úÖ

    Non-negative, discrete input features (like counts) ‚úÖ


If a feature and the target are dependent, the chi-square score will be high, meaning the feature is important.

If they are independent, the score will be low (or zero), so we can drop the feature.

````python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.datasets import load_iris
import pandas as pd

# Load a sample dataset (Iris)
data = load_iris()
X = data.data         # Features
y = data.target       # Target

# Convert to DataFrame (for better feature naming)
df = pd.DataFrame(X, columns=data.feature_names)
print("Original Features:\n", df.head())

# Apply chi-squared test to select top 2 features
chi_selector = SelectKBest(score_func=chi2, k=2)
X_kbest = chi_selector.fit_transform(X, y)

# Get selected feature names
mask = chi_selector.get_support()  # Boolean mask of selected features
selected_features = df.columns[mask]

print("\nSelected Features (Top 2 by Chi¬≤):", list(selected_features))
````

### Box-Cox Transformation

We can see that the non-normal distribution was converted into a normal distribution or rather close to normal using 

The Box-Cox transformation is a power transformation used to make data more normally distributed (i.e., bell-shaped). It is especially useful when your data is skewed (not symmetric), which can negatively affect models that assume normality

````python
# Python3 code to show Box-cox Transformation 
# of non-normal data

# import modules
import numpy as np
from scipy import stats

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt

# generate non-normal data (exponential)
original_data = np.random.exponential(size = 1000)

# transform training data & save lambda value
fitted_data, fitted_lambda = stats.boxcox(original_data)

# creating axes to draw plots
fig, ax = plt.subplots(1, 2)

# plotting the original data(non-normal) and 
# fitted data (normal)
sns.distplot(original_data, hist = False, kde = True,
			kde_kws = {'shade': True, 'linewidth': 2}, 
			label = "Non-Normal", color ="green", ax = ax[0])

sns.distplot(fitted_data, hist = False, kde = True,
			kde_kws = {'shade': True, 'linewidth': 2}, 
			label = "Normal", color ="green", ax = ax[1])

# adding legends to the subplots
plt.legend(loc = "upper right")

# rescaling the subplots
fig.set_figheight(5)
fig.set_figwidth(10)

print(f"Lambda value used for Transformation: {fitted_lambda}")
````

### Class weights

Many machine learning models allow you to assign more importance to the minority class by using class weights. This prevents the model from being biased toward the majority class by penalizing it more heavily when it misclassifies minority samples. This is particularly useful in: Logistic Regression, SVM, Decision Trees, Random Forest

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification


# Sample imbalanced data
X, y = make_classification(n_samples=1000, n_classes=2, weights=[0.9, 0.1], random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2)

# Use class_weight='balanced'
model = RandomForestClassifier(class_weight='balanced', random_state=42)
model.fit(X_train, y_train)
````

### Evaluation Metrics

In imbalanced datasets, accuracy is misleading. Instead, we use:
‚úÖ Precision, Recall, F1-score:

    Precision: How many predicted positives are actually positive?

    Recall: How many actual positives were correctly predicted?

    F1-score: Harmonic mean of precision and recall.

‚úÖ ROC-AUC:

    Measures how well the model separates classes at various thresholds.

    Good when classes are not heavily imbalanced.

‚úÖ PR-AUC (Precision-Recall AUC):

    Better for very imbalanced datasets.

    Focuses more on the minority class.

````python
from sklearn.metrics import classification_report, roc_auc_score, precision_recall_curve, auc
import matplotlib.pyplot as plt

# Predict labels and probabilities
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

# 1. Precision, Recall, F1
print(classification_report(y_test, y_pred))

# 2. ROC-AUC
roc_auc = roc_auc_score(y_test, y_proba)
print(f"ROC-AUC: {roc_auc:.4f}")

# 3. PR-AUC
precision, recall, _ = precision_recall_curve(y_test, y_proba)
pr_auc = auc(recall, precision)
print(f"PR-AUC: {pr_auc:.4f}")

# Plot PR Curve
plt.figure()
plt.plot(recall, precision, label=f'PR AUC = {pr_auc:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid(True)
plt.show()
````

### PCA

Principal component analysis is a method that we would like to get the important features by reducing the dimensionality and get the most important features. This term would set some component to the data and try to reduce the dimension as the data features tends to infinity.

### How to ‚ÄúChoose‚Äù the Number of Principal Components (PCs) Wisely

When you run PCA on a dataset, you start with *n* original features (dimensions). PCA computes up to *n* principal components, each of which is a new ‚Äúaxis‚Äù (a linear combination of the old features) that captures as much of the data‚Äôs variance as possible, in descending order.

- **PC1** is the direction that captures the most variance.  
- **PC2** is orthogonal to PC1 and captures the next‚Äêlargest chunk of variance.  
- And so on, down to **PCn**.

But in practice, you rarely need all *n* components. Why? Because after a point, adding more PCs only ‚Äúmops up‚Äù very tiny bits of variance (noise). You want to keep as few PCs as possible while still preserving *most* of the important information in the data.

---

#### The Trade‚ÄêOff

- Too few PCs ‚Üí you lose important structure (under‚Äêrepresent your data).  
- Too many PCs ‚Üí you defeat the purpose of dimensionality reduction (you end up almost where you started).

So the question is:

> ‚ÄúHow many PCs are **enough** for my purposes?‚Äù

That depends on:

1. **How much of the total variance** you feel comfortable ‚Äúkeeping.‚Äù  
2. **What your downstream task is** (e.g. visualization, clustering, classification, etc.).  
3. **How many dimensions** you can handle computationally or for interpretability.  

---

#### What ‚ÄúExplained Variance Ratio‚Äù Actually Means

Imagine your dataset as a collection of points in 4D (for Iris, 4 features). If you measure how ‚Äúspread out‚Äù the data is along each original feature, you get some sense of variability in each dimension. PCA does something like this:

1. **Find PC1**: the single direction in 4D that has the largest spread of all the data.  
   - If you were to project every point onto that line, you would capture the maximum possible ‚Äúspread‚Äù compared to any other line through the origin.  
   - The *explained variance* of PC1 is simply ‚Äúhow much of the total data variance lives along that line.‚Äù

2. **Find PC2**: the next direction, orthogonal to PC1, that captures the next‚Äêlargest chunk of variance.  
3. And so on.

When you do:
```python
pca = PCA().fit(scaled_data)
print(pca.explained_variance_ratio_)
```
you get an array like (for Iris, approximate numbers):
```
[0.7209,  0.2289,  0.0473,  0.0030]
```
- That means **PC1** captures about **72.09%** of the total variance in the 4D data.  
- **PC2** captures about **22.89%**.  
- **PC3** captures about **4.73%**.  
- **PC4** captures about **0.30%**.

If you sum them cumulatively:
```python
cumulative = np.cumsum([0.7209, 0.2289, 0.0473, 0.0030])
# cumulative ‚Üí [0.7209, 0.9498, 0.9971, 1.0000]
```
- **After PC1**, you already have **72.1%** of the variance.  
- **After PC1 + PC2**, you have **94.98%** of the variance (‚âà95%).  
- **After PC1 + PC2 + PC3**, you have **99.71%**.  
- And the last PC brings it up to **100%** (in theory).

---

#### Using a Threshold: Why 0.95, and Can It Be Dynamic?

You saw in your plot that the ‚Äúred dashed line‚Äù was at **0.95 (95%)**. That‚Äôs just a rule of thumb‚Äîpeople often say ‚Äúkeep enough PCs to explain at least 90% or 95% (or 99%) of the variance, depending on how much information you‚Äôre willing to lose.‚Äù

#### Why 0.95?

- **95%** is a compromise: you keep almost all of the structure (information) but drop the last 5% of ‚Äútiny wiggles.‚Äù  
- If **PC1** already gives you **95.58%**, that means a single axis captures nearly all of the data‚Äôs variability. You could reduce from 4D ‚Üí 1D and still be pretty faithful.

But that ‚Äú95%‚Äù is **not sacred**. You could choose:

- **0.90 (90%)** if you want a more aggressive reduction.  
- **0.99 (99%)** if you need near‚Äêperfect retention.  
- Or something else entirely, depending on your problem.

---

#### How to Pick a Threshold Dynamically

1. **Look at the cumulative curve**: Sometimes it has a clear ‚Äúelbow.‚Äù You might choose the number of PCs at that elbow (where adding more PCs mostly flattens out).  
2. **Consider downstream performance**:  
   ```python
   for k in range(1, n_features+1):
       X_transformed = PCA(n_components=k).fit_transform(X_original)
       score = cross_val_score(my_classifier, X_transformed, y, cv=5).mean()
       if score >= desired_accuracy:
           print("Choose k=", k)
           break
   ```
   - Try 1 PC, 2 PCs, 3 PCs, etc., see which gives you the best performance vs. complexity trade‚Äêoff.  
3. **User‚Äêdefined error tolerance**: ‚ÄúI only want to lose at most 5% of variance‚Äù ‚Üí `threshold = 0.95`.  
4. **Computational constraints**: If training with 4 features is too slow, but 2 features is fast enough, see if 2 PCs meet your variance goal.  

In short, the threshold (`0.95`, `0.90`, `0.99`, etc.) should be chosen based on:

- **How much information you‚Äôre willing to ‚Äúthrow away.‚Äù**  
- **The requirements of the task you‚Äôll do next** (visualization, clustering, classification, etc.).  
- **The computational budget or desire for interpretability.**

---

#### Putting It All Together: A Dynamic, Annotated Example

Below is a self‚Äêcontained script that:

1. Loads Iris  
2. Scales the data  
3. Fits a full PCA  
4. Prints out both the individual explained‚Äêvariance ratios **and** the cumulative sums  
5. Shows how you could dynamically pick the number of PCs by inspecting a ‚Äútarget variance.‚Äù


---

#### What This Code Does, Line by Line

1. **Load & scale** your 4‚Äêdimensional Iris data.  
2. **Fit** a full PCA (all PCs).  
3. **Print** each PC‚Äôs ‚Äúexplained variance ratio‚Äù (e.g. `[0.7209, 0.2289, 0.0473, 0.0030]`):  
   - PC1 explains ~72.09%.  
   - PC2 explains ~22.89%.  
   - PC3 explains ~4.73%.  
   - PC4 explains ~0.30%.  
4. **Compute cumulative sum** of those ratios:  
   - Cumulative[0] = 0.7209  
   - Cumulative[1] = 0.7209 + 0.2289 = 0.9498 (~94.98%)  
   - Cumulative[2] = 0.9971 (~99.71%)  
   - Cumulative[3] = 1.0000 (100%)  
5. **Plot** those cumulative values versus ‚Äú# of PCs.‚Äù The red and green dashed lines at 95% and 90% show where your curve crosses common thresholds.  
6. **Find** `num_pcs = argmax(cumulative >= 0.95) + 1`. This picks the **smallest k** such that cumulative variance ‚â• 0.95. In this case, it should print `k = 2`.  
7. **Print** how many PCs are needed for 90%, 95%, and 99% variance.

---

#### All main questions

1. **‚ÄúHow could we choose the number of PCs wisely?‚Äù**  
   - **Look at the cumulative explained variance curve.** Decide on a threshold that makes sense (e.g. 90% or 95%). Pick the smallest *k* where cumulative ‚â• your threshold.  
   - **Alternatively**, use a ‚Äúcross‚Äêvalidation approach‚Äù in your downstream model: try 1 PC, 2 PCs, 3 PCs, etc., see which gives you the best performance vs. complexity trade‚Äêoff.  
   - **Or look for an ‚Äúelbow‚Äù**‚Äîthe point where adding one more PC barely moves the curve up.

2. **‚ÄúWhat does the explained_variance_ratio_ actually do?‚Äù**  
   - It tells you **what fraction of the total dataset variance** is captured by each PC.  
   - Example: if `explained_variance_ratio_ = [0.72, 0.23, 0.04, 0.01]`, then PC1 explains 72% of the variability, PC2 explains another 23%, and so on.  
   - By summing them, you know ‚Äúif I keep PC1 and PC2, I‚Äôve captured 72% + 23% = 95% of the original structure.‚Äù

3. **‚ÄúIn the chart, why is 0.95 chosen as the threshold? Can it be dynamic, not static?‚Äù**  
   - **0.95 is just a common rule of thumb.** It means ‚ÄúI am okay throwing away 5% of the variance (noise) to reduce dimensions drastically.‚Äù  
   - You can choose **any threshold** you like‚Äîeven 0.80 for a very aggressive reduction, or 0.99 if you need nearly perfect reconstruction.  
   - To be dynamic, you usually:  
     1. **Compute** the cumulative explained variance for all PCs.  
     2. **Pick** the smallest *k* such that `cumulative[k‚àí1] ‚â• desired_threshold`.  
   - In code, that‚Äôs exactly what `np.argmax(cumulative >= threshold) + 1` does.

---


#### Final Recommendations

1. **Compute `explained_variance_ratio_` and its cumulative sum** every time you run PCA.  
2. **Plot the cumulative curve** if you want a visual ‚Äúelbow‚Äù check.  
3. **Choose a threshold** (0.80, 0.90, 0.95, 0.99) based on how much information you‚Äôre willing to sacrifice.  
4. **Select the smallest *k*** such that `cumulative_variance[k‚àí1] ‚â• threshold`.  
5. If you have a downstream task (e.g. classification), you can also **cross‚Äêvalidate directly** on different *k* to see what yields the best performance.

---

üöÄ **One‚ÄêLine Code to Pick `n_components` Dynamically for 95%:**


üéØ **Key Takeaway**

- ‚ÄúExplained variance ratio‚Äù tells you, for each PC, **‚Äúhow much of the data‚Äôs total variance does this PC cover.‚Äù**  
- ‚ÄúCumulative explained variance‚Äù sums that up: **‚Äúif I keep PC1 through PC*k*, I‚Äôm capturing**  
  \(\sum_{i=1}^k \text{explained_variance_ratio}_i\).  
- **Pick the fewest PCs** such that that cumulative sum meets or exceeds your chosen threshold (e.g. 0.95), **or** pick the *k* that gives the best downstream results if you‚Äôre optimizing for classification/regression performance.



````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

def best_num_pcs(data_scaled: np.ndarray, target_variance: float = 0.95) -> int:
    """
    Given a standardized dataset `data_scaled` and a desired cumulative explained-variance
    threshold `target_variance` (e.g. 0.95 for 95%), this function fits a full PCA and
    returns the smallest number of principal components required to reach at least
    `target_variance` of the total variance.
    
    Parameters:
    -----------
    data_scaled : np.ndarray
        The standardized (zero-mean, unit-variance) input data, shape (n_samples, n_features).
    target_variance : float (default=0.95)
        The cumulative explained-variance threshold (between 0 and 1).
        
    Returns:
    --------
    num_components : int
        The minimum number of principal components needed to reach ‚â• target_variance.
    """
    # 1. Fit a full PCA (no limit on n_components)
    pca_full = PCA().fit(data_scaled)
    
    # 2. Extract explained_variance_ratio_ and compute its cumulative sum
    explained = pca_full.explained_variance_ratio_
    cumulative = np.cumsum(explained)
    
    # 3. Find the smallest index where cumulative ‚â• target_variance
    #    +1 because indices are zero-based but component count is 1-based
    num_components = np.argmax(cumulative >= target_variance) + 1
    return num_components


# -------------------------------
#          MAIN SCRIPT
# -------------------------------

# 1. Load the Iris dataset
iris = datasets.load_iris()
X = iris.data        # shape: (150, 4)
y = iris.target      # shape: (150,)

# 2. Convert to DataFrame (optional, but useful if you want to inspect column names)
df = pd.DataFrame(X, columns=iris.feature_names)

# 3. Standardize the features (mean=0, std=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# 4. Use our function `best_num_pcs` to determine how many PCs we need for ‚â• 95% variance
threshold = 0.95
n_optimal = best_num_pcs(X_scaled, target_variance=threshold)
print(f"Optimal number of PCs to reach ‚â• {threshold*100:.0f}% variance: {n_optimal}\n")

# 5. Fit PCA with the optimal number of components
pca_opt = PCA(n_components=n_optimal)
X_pca = pca_opt.fit_transform(X_scaled)

# 6. Create a DataFrame for PCA-transformed data
pc_column_names = [f"PC{i+1}" for i in range(n_optimal)]
pca_df = pd.DataFrame(X_pca, columns=pc_column_names)

# 7. For completeness, print out the explained_variance_ratio_ and cumulative sums
explained_ratios = pca_opt.explained_variance_ratio_
cumulative_explained = np.cumsum(explained_ratios)

print("Explained variance ratio of each selected PC:")
for idx, ratio in enumerate(explained_ratios, start=1):
    print(f"  PC{idx}: {ratio:.4f}  (‚âà {ratio*100:.2f}%)")
print("\nCumulative explained variance (selected PCs):")
for idx, cum_val in enumerate(cumulative_explained, start=1):
    print(f"  PC1‚ÜíPC{idx}: {cum_val:.4f}  (‚âà {cum_val*100:.2f}%)")
print()

# 8. Plot the cumulative explained variance curve for all PCs (to visualize elbow & thresholds)
#    (This uses a fresh ‚Äúfull PCA‚Äù fit just for plotting purposes.)
pca_full_for_plot = PCA().fit(X_scaled)
explained_full = pca_full_for_plot.explained_variance_ratio_
cumulative_full = np.cumsum(explained_full)

plt.figure(figsize=(7, 4))
plt.plot(
    np.arange(1, len(cumulative_full) + 1),
    cumulative_full,
    marker='o',
    linestyle='-',
    label='Cumulative Explained Variance'
)
plt.axhline(y=threshold, color='r', linestyle='--', label=f'{int(threshold*100)}% Threshold')
plt.title("Cumulative Explained Variance vs. Number of Principal Components")
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.xticks(np.arange(1, len(cumulative_full) + 1))
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()

# 9. Visualize the PCA-transformed data in 2D if at least 2 PCs were chosen
if n_optimal >= 2:
    plt.figure(figsize=(6, 5))
    sns.scatterplot(
        data=pca_df,
        x='PC1',
        y='PC2',
        hue=y,
        palette='Set2',
        edgecolor='k',
        s=60
    )
    plt.title(f"Iris projected onto first 2 PCs (n_components={n_optimal})")
    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.legend(title="Species", labels=iris.target_names)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
else:
    # If only 1 PC was chosen, show a simple histogram of PC1
    plt.figure(figsize=(6, 4))
    sns.histplot(pca_df['PC1'], bins=20, kde=True, color='skyblue')
    plt.title(f"Iris projected onto PC1 only (n_components={n_optimal})")
    plt.xlabel("Principal Component 1")
    plt.ylabel("Frequency")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
````

# KNN

KNN is a NON-Parameteric (without any loss function) captures the idea of similarity (sometimes called distance, proximity, or closeness) with calculating the distance between points on a graph

Euclidean distance
is a popular and familiar choice. It is widely disposable in real-life scenarios since it is non-parametric, meaning, it does not make any underlying assumptions about the distribution of data (as opposed to other algorithms such as GMM, which assume a Gaussian distribution of the given data) 

What This Code Does:
- load_and_prepare_data()-> Loads and scales the Iris dataset (standard or min-max scaling)
- select_features_chi2()-> Applies Chi-squared feature selection to pick top k features
- elbow_method() -> Tests k values from 1‚Äì20 and finds the one with highest test accuracy
- cross_val_best_k() ->	Uses 5-fold cross-validation to find the best k
- evaluate_model() ->	Shows classification report and confusion matrix for best model
- Final step -> You train and test KNN with best k based on CV

Although this is just a reference. because the StandardScaler and normalizations all of them make the X to be negative which cannot be.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    ConfusionMatrixDisplay,
)

# ============================== #
#      DATA PREP + SCALING       #
# ============================== #

def load_and_prepare_data(scale_method='standard'):
    data = load_iris()
    X = data.data
    y = data.target

    # Scale: 'standard' or 'minmax'
    if scale_method == 'standard':
        scaler = StandardScaler()
    elif scale_method == 'minmax':
        scaler = MinMaxScaler()
    else:
        raise ValueError("scale_method must be 'standard' or 'minmax'")

    X_scaled = scaler.fit_transform(X)

    return X_scaled, y, data.target_names

# ============================== #
#       FEATURE SELECTION        #
# ============================== #

def select_features_chi2(X, y, k=2):
    selector = SelectKBest(score_func=chi2, k=k)
    X_kbest = selector.fit_transform(X, y)
    return X_kbest

# ============================== #
#     ELBOW METHOD FOR BEST K    #
# ============================== #

def elbow_method(X_train, y_train, X_test, y_test, max_k=20):
    accuracies = []
    k_range = range(1, max_k + 1)
    
    for k in k_range:
        model = KNeighborsClassifier(n_neighbors=k)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        accuracies.append(acc)
    
    best_k = k_range[np.argmax(accuracies)]
    best_accuracy = max(accuracies)

    # Plot
    plt.figure(figsize=(8, 5))
    plt.plot(k_range, accuracies, marker='o')
    plt.title("Elbow Method: KNN Accuracy vs K")
    plt.xlabel("K")
    plt.ylabel("Accuracy")
    plt.grid(True)
    plt.xticks(k_range)
    plt.show()

    return best_k, best_accuracy

# ============================== #
#     CROSS-VALIDATION METHOD    #
# ============================== #

def cross_val_best_k(X, y, max_k=20, cv=5):
    mean_scores = []
    k_range = range(1, max_k + 1)

    for k in k_range:
        model = KNeighborsClassifier(n_neighbors=k)
        scores = cross_val_score(model, X, y, cv=cv)
        mean_scores.append(scores.mean())

    best_k = k_range[np.argmax(mean_scores)]
    best_accuracy = max(mean_scores)

    # Plot
    plt.figure(figsize=(8, 5))
    plt.plot(k_range, mean_scores, marker='s', color='green')
    plt.title("Cross-Validation: Accuracy vs K")
    plt.xlabel("K")
    plt.ylabel("CV Accuracy")
    plt.grid(True)
    plt.xticks(k_range)
    plt.show()

    return best_k, best_accuracy

# ============================== #
#           EVALUATION           #
# ============================== #

def evaluate_model(model, X_test, y_test, target_names, best_k):
    y_pred = model.predict(X_test)

    print("\nüìä Classification Report:\n", classification_report(y_test, y_pred))
    print("üß© Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

    ConfusionMatrixDisplay.from_predictions(y_test, y_pred, display_labels=target_names)
    plt.title(f"Confusion Matrix (Best K = {best_k})")
    plt.grid(False)
    plt.show()

# ============================== #
#             MAIN               #
# ============================== #

if __name__ == "__main__":
    # Step 1: Load & scale data (choose 'standard' or 'minmax')
    X, y, target_names = load_and_prepare_data(scale_method='minmax')

    # Step 2: Feature Selection
    X_kbest = select_features_chi2(X, y, k=2)

    # Step 3: Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X_kbest, y, test_size=0.2, random_state=42
    )

    # Step 4: Find best K using Elbow
    print("\nüîç Finding best K using Elbow Method...")
    best_k_elbow, acc_elbow = elbow_method(X_train, y_train, X_test, y_test)

    # Step 5: Find best K using Cross-Validation
    print("\nüî¨ Finding best K using Cross-Validation...")
    best_k_cv, acc_cv = cross_val_best_k(X_kbest, y, cv=5)

    print(f"\n‚úÖ Best K (Elbow): {best_k_elbow} ‚Üí Accuracy: {round(acc_elbow, 3)}")
    print(f"‚úÖ Best K (CV): {best_k_cv} ‚Üí Accuracy: {round(acc_cv, 3)}")

    # Step 6: Train final model using best K (from cross-validation)
    final_k = best_k_cv
    model = KNeighborsClassifier(n_neighbors=final_k, p=2)
    model.fit(X_train, y_train)

    # Step 7: Evaluate final model
    evaluate_model(model, X_test, y_test, target_names, final_k)
````

2. Split into train/test
3. Cross-validation on training set
4. Hyperparameter tuning (GridSearchCV)

6. Test model on test set
7. Evaluate: accuracy, confusion matrix, etc.



````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    ConfusionMatrixDisplay,
)

# ======================= #
# 1. Load and Prepare Data
# ======================= #
iris = load_iris()
X, y = iris.data, iris.target
target_names = iris.target_names

# Scale features (important for KNN!)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into Train / Test (80/20)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# ================================ #
# 2. Define Hyperparameter Grid
# ================================ #
param_grid = {
    "n_neighbors": list(range(1, 31)),
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan"]
}

# ================================ #
# 3. Setup GridSearchCV with CV=5
# ================================ #
knn = KNeighborsClassifier()

grid = GridSearchCV(
    estimator=knn,
    param_grid=param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1,
    verbose=1
)

grid.fit(X_train, y_train)

# ======================= #
# 4. Show Best Parameters
# ======================= #
print("‚úÖ Best Hyperparameters:", grid.best_params_)
print("üìà Best CV Accuracy:     ", round(grid.best_score_, 4))

# ============================= #
# 5. Evaluate on Test Set
# ============================= #
best_model = grid.best_estimator_

y_pred = best_model.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print("üß™ Test Accuracy:", round(acc, 4))
print("\nüìä Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, display_labels=target_names)
plt.title("Confusion Matrix (KNN)")
plt.grid(False)
plt.show()
````

## Decision Trees

A decision tree is a supervised learning algorithm used for both classification and regression tasks. It models decisions as a tree-like structure where internal nodes represent attribute tests, branches represent attribute values and leaf nodes represent final decisions or predictions. Decision trees are versatile, interpretable, and widely used in machine learning for predictive modeling.

## Decision Tree Classifier

In order to make the classifieds better we need to gain some info. these info regarding their mixed up situation and uncertainity would be splitted into 2 categories. each categories would represent the fact and the concept of reducing the mixed up or messy or disorderity by SPLITTING THE DATA. 

<b>‚ÄúLet‚Äôs see how pure each piece is after the cut.‚Äù</b>

1. Information gain(Entropy): Measures the amount of <b>UNCERTAINITY</b> in your data.


2. Gini Index: Measures how often you‚Äôd be wrong if you picked a label at random.

### Gini Index

$$
Gini(p) = 1 - \sum_{i=1}^{n} p_i^2
$$

### Entropy

$$
Entropy(p) = - \sum_{i=1}^{n} p_i \log_2 p_i
$$



## Hyperparameter Tuning with Decision Tree Classifier

Hyperparameters are configuration settings that control the behavior of a decision tree model and significantly affect its performance. Proper tuning can improve accuracy, reduce overfitting and enhance generalization of model.

Popular methods for tuning include Grid Search, Random Search, and Bayesian Optimization, which explore different combinations to find the best configuration.

### RandomizedSearchCV

Imagine you have a recipe and you want to make it perfect. You try 10 different combinations of salt, sugar, and spices ‚Äî randomly. Maybe some work well, some don‚Äôt. That‚Äôs what RandomizedSearchCV does.

Instead of trying every single combo (like Grid Search), it picks random combinations of settings and tests them. It‚Äôs like gambling ‚Äî fast, and sometimes lucky.
it‚Äôs much faster than testing everything. Especially when you have a lot of settings (like depth of tree, number of leaves, etc.).

If you have:

    max_depth = [3, 4, 5]

    criterion = ["gini", "entropy"]

    min_samples_leaf = [1, 2, 3]

Then total combinations =
3√ó2√ó3=18
3√ó2√ó3=18

RandomizedSearchCV lets you say:
‚ÄúI‚Äôll try only 5 random ones out of these 18.‚Äù

### Optuna

Now imagine instead of guessing recipes randomly, you have a smart chef. You try one recipe, he tastes it and says:

    "Hmm, too salty. Try a little less salt and a bit more sugar."

That‚Äôs what Optuna does. It‚Äôs a smart tuner. It learns from each trial and guesses better next time. It uses a strategy called Bayesian optimization, but don‚Äôt let the name scare you.


Because it‚Äôs smarter than RandomSearch ‚Äî it finds good settings faster, with fewer tries.

### ASHA / Hyperband


Let‚Äôs go back to the recipe story. You‚Äôre trying 20 new recipes. Instead of fully cooking all of them, you taste each after 5 minutes, and if one already tastes bad ‚Äî you throw it away early.

That‚Äôs what ASHA (or Hyperband) does. It doesn‚Äôt waste time on bad models. It stops them early and only continues training the promising ones.

It‚Äôs like:

    ‚ÄúIf a model is bad in the first 10 minutes, why waste 1 hour on it?‚Äù

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import accuracy_score
import optuna
from optuna.pruners import SuccessiveHalvingPruner

# -- 1) Prepare data --
data = {
    "Color": ["Red","Red","Green","Blue","Blue","Green","Red","Green","Blue","Red"],
    "Type":  ["Gummy","Chocolate","Hard","Hard","Gummy","Chocolate","Hard","Gummy","Chocolate","Hard"],
    "Liked": ["Yes","Yes","No","No","Yes","Yes","No","Yes","Yes","No"]
}
df = pd.DataFrame(data)
X = pd.get_dummies(df[["Color","Type"]])
y = df["Liked"].map({"Yes":1,"No":0})
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# -- 2) Default trees --
gini_tree    = DecisionTreeClassifier(criterion="gini",    random_state=0).fit(X_train, y_train)
entropy_tree = DecisionTreeClassifier(criterion="entropy", random_state=0).fit(X_train, y_train)
acc_g  = accuracy_score(y_test, gini_tree.predict(X_test))
acc_e  = accuracy_score(y_test, entropy_tree.predict(X_test))

# -- 3) RandomizedSearchCV (very small grid) --
param_grid = {
    "criterion":       ["gini","entropy"],
    "max_depth":       [2,3],
    "min_samples_leaf":[1,2]
}
rs = RandomizedSearchCV(
    DecisionTreeClassifier(random_state=0),
    param_distributions=param_grid,
    n_iter=3, cv=2, random_state=42
).fit(X_train, y_train)
rs_tree = rs.best_estimator_
acc_rs  = accuracy_score(y_test, rs_tree.predict(X_test))

# -- 4) Optuna (light) --
def objective(trial):
    params = {
        "criterion": trial.suggest_categorical("criterion", ["gini","entropy"]),
        "max_depth": trial.suggest_int("max_depth", 2, 3),
        "min_samples_leaf": trial.suggest_int("min_samples_leaf", 1, 2)
    }
    m = DecisionTreeClassifier(**params, random_state=0)
    m.fit(X_train, y_train)
    return accuracy_score(y_test, m.predict(X_test))

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=5)
opt_tree = DecisionTreeClassifier(**study.best_params, random_state=0).fit(X_train, y_train)
acc_opt  = accuracy_score(y_test, opt_tree.predict(X_test))

# -- 5) Optuna + ASHA (Hyperband-style) --
study2 = optuna.create_study(direction="maximize", pruner=SuccessiveHalvingPruner())
study2.optimize(objective, n_trials=5)
asha_tree = DecisionTreeClassifier(**study2.best_params, random_state=0).fit(X_train, y_train)
acc_asha  = accuracy_score(y_test, asha_tree.predict(X_test))

# -- 6) Plot all results without axs variables --
# Default Gini vs Entropy
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plot_tree(gini_tree, feature_names=X.columns, class_names=["No","Yes"], filled=True)
plt.title(f"Gini (Acc {acc_g:.2f})")
plt.subplot(1,2,2)
plot_tree(entropy_tree, feature_names=X.columns, class_names=["No","Yes"], filled=True)
plt.title(f"Entropy (Acc {acc_e:.2f})")
plt.tight_layout()
plt.show()

# Tuned models: Random, Optuna, ASHA
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plot_tree(rs_tree, feature_names=X.columns, class_names=["No","Yes"], filled=True)
plt.title(f"RandomSearch (Acc {acc_rs:.2f})")
plt.subplot(1,3,2)
plot_tree(opt_tree, feature_names=X.columns, class_names=["No","Yes"], filled=True)
plt.title(f"Optuna (Acc {acc_opt:.2f})")
plt.subplot(1,3,3)
plot_tree(asha_tree, feature_names=X.columns, class_names=["No","Yes"], filled=True)
plt.title(f"ASHA (Acc {acc_asha:.2f})")
plt.tight_layout()
plt.show()
````

## Decision Tree Regression

Decision Tree Regression is a powerful and intuitive machine learning technique. Unlike traditional linear regression, which assumes a straight-line relationship between input features and the target variable, Decision Tree Regression is a non-linear regression method that can handle complex datasets with intricate patterns. It uses a tree-like model to make predictions, making it both flexible and easy to interpret.

Decision Tree Regression predicts continuous values. It does this by splitting the data into smaller subsets based on decision rules derived from the input features. Each split is made to minimize the error in predicting the target variable. At leaf node of the tree the model predicts a continuous value which is typically the average of the target values in that node.

The branches or edges represent the result of the node and the nodes have either: 

Conditions (Decision Nodes): The diamond-shaped nodes represent conditions or decision points. They ask a question about the input data (e.g., ‚ÄúIs feature X greater than 5?‚Äù).
Result (leaf Nodes): The rectangular boxes represent the final outcomes or results. In regression, the leaf node contains the predicted continuous value.

````python
from sklearn.tree import plot_tree
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, export_text
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Generate synthetic dataset
np.random.seed(42)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])

# Visualize the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

reg = DecisionTreeRegressor(max_depth=4, random_state=42)
reg.fit(X_train,y_train)

pred = reg.predict(X_test)
test = mean_squared_error(X_test, pred)

X_grid = np.arange(min(X), max(X), 0.01)[:, np.newaxis]
y_grid_pred = reg.predict(X_grid)
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='red', label='Data')
plt.plot(X_grid, y_grid_pred, color='blue', label='Model Prediction')
plt.title("Decision Tree Regression")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()
plt.show()

plt.figure(figsize=(20, 10))
plot_tree(
    reg,
    feature_names=["Feature"],
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree Structure")
plt.show()
````
